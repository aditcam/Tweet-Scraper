```python
#checking if the tweet had photos 
img = []
for idx,tweet in enumerate(all_tweets):
    tweet_media = str(tweet.entities) 
    if "photo" in tweet_media: 
        img.append('True')
        
    else: 
        img.append('None') 
        
print(len(img))
print(img)
```

```python
#write output to json files 
with open("output.jsonl",'w') as outfile:
     for entry in outtweets:
         json.dump(entry,outfile,default=str)
         outfile.write('\n') 
```
